```python
import pandas as pd
original  = pd.read_csv('df_Mecab2Mecab.csv', encoding = "utf8")
df = original.dropna(axis = 0).reset_index()
del(df['index'])
df.columns = ['text']

list1 = []
for sen in df['text']:
    for vocab in sen.split():
        list1.append(vocab)
        
set1 = set(list1)
len(set1)
```
