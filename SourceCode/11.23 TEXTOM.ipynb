## 시각화 - TEXTOM: Mecab

```python
import pandas as pd
df_TextomEK = pd.read_csv("TEXTOM_ZIPUP_Mecab.txt", sep = "\n", engine='python', encoding = "utf8",header=None)
df_TextomEK
temp = df_TextomEK.values.tolist()
list_TextomEK = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_TextomEK.append(vocab)
series_TextomEK = pd.Series(list_TextomEK)
TextomEK_counts = series_TextomEK.value_counts()
TextomEK_counts
```

```python
from matplotlib import pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rc('font',family = "Malgun Gothic")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
TextomEK_counts.head(50).plot(figsize=(10,15),kind='barh',grid=True, title='Textom Mecab')
```
