## Mecab

```python
import pandas as pd
original  = pd.read_csv('df_Mecab.csv', encoding = "utf8")
df = original.dropna(axis = 0).reset_index()
del(df['index'])
df.columns = ['text']

list1 = []
for sen in df['text']:
    for vocab in sen.split():
        list1.append(vocab)
        
set1 = set(list1)
len(set1)
```

## 교집합 단어별 빈도 수

```python
import pandas as pd
```

```python
df_Hannanum = pd.read_csv("df_Hannanum.csv").dropna()

temp = df_Hannanum.values.tolist()
list_Hannanum = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_Hannanum.append(vocab)
series_Hannanum = pd.Series(list_Hannanum)
Hannanum_counts = series_Hannanum.value_counts()
Hannanum_counts
```

```python
df_Kkma = pd.read_csv("df_Kkma.csv").dropna()

temp = df_Kkma.values.tolist()
list_Kkma = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_Kkma.append(vocab)
series_Kkma = pd.Series(list_Kkma)
Kkma_counts = series_Kkma.value_counts()
Kkma_counts
```

```python
df_Komoran = pd.read_csv("df_Komoran.csv").dropna()

temp = df_Komoran.values.tolist()
list_Komoran = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_Komoran.append(vocab)
series_Komoran = pd.Series(list_Komoran)
Komoran_counts = series_Komoran.value_counts()
Komoran_counts
```

```python
df_Mecab = pd.read_csv("df_Mecab.csv").dropna()

temp = df_Mecab.values.tolist()
list_Mecab = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_Mecab.append(vocab)
series_Mecab = pd.Series(list_Mecab)
Mecab_counts = series_Mecab.value_counts()
Mecab_counts
```

```python
df_Okt = pd.read_csv("df_Okt.csv").dropna()

temp = df_Okt.values.tolist()
list_Okt = []
for sentence in temp:
    temp2 = sentence[0].split()
    for vocab in temp2:
        list_Okt.append(vocab)
series_Okt = pd.Series(list_Okt)
Okt_counts = series_Okt.value_counts()
Okt_counts
```

```python
vocab_freq = pd.DataFrame(intersection)
vocab_freq = vocab_freq.rename(columns={0:'vocab'})
vocab_freq
```

```python
vocab_freq[['Hannanum', 'Kkma', 'Komoran', 'Mecab', 'Okt']] = 0
```

```python
vocab_freq
```

```python
vocab_freq['vocab'][0]
```

```python
Mecab_counts['구명']
```

```python
for i in range(len(vocab_freq)):
    word = vocab_freq['vocab'][i]
    vocab_freq['Hannanum'][i] = Hannanum_counts[word]
vocab_freq
```

```python
for i in range(len(vocab_freq)):
    word = vocab_freq['vocab'][i]
    vocab_freq['Kkma'][i] = Kkma_counts[word]
    
for i in range(len(vocab_freq)):
    word = vocab_freq['vocab'][i]
    vocab_freq['Komoran'][i] = Komoran_counts[word]
    
for i in range(len(vocab_freq)):
    word = vocab_freq['vocab'][i]
    vocab_freq['Mecab'][i] = Mecab_counts[word]
    
for i in range(len(vocab_freq)):
    word = vocab_freq['vocab'][i]
    vocab_freq['Okt'][i] = Okt_counts[word]
    
vocab_freq
```

```python
vocab_freq['Total'] = 0
vocab_freq
```

```python
for i in range(len(vocab_freq)):
    vocab_freq['Total'][i] = vocab_freq['Hannanum'][i] + vocab_freq['Kkma'][i] 
    + vocab_freq['Komoran'][i] + vocab_freq['Mecab'][i] + vocab_freq['Okt'][i]
```
